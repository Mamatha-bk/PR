## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`. 

1. First, create a numpy array that contains the numbers $0,1, \cdots, N$ where $N$ is the number of rows in the image.

2. Use `numpy.random.shuffle` to shuffle the indices. Call the shuffled array, `rows`

4. Print `rows` and inspect the elements.

```python
import numpy as np

# Assuming img_average_channel is a NumPy array
num_rows = img_average_channel.shape[0]

# Create an array of row indices from 0 to num_rows-1
rows = np.arange(num_rows)

# Shuffle the row indices using numpy.random.shuffle
np.random.shuffle(rows)

# Print the shuffled row indices
print("Shuffled Row Indices:")
print(rows)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
# Create matrix T by rearranging the rows of identity_matrix
T = identity_matrix[rows]

# Ensure that T has the same shape as identity_matrix and img_average_channel
if T.shape != identity_matrix.shape:
    T = T.reshape(identity_matrix.shape)

# Sanity check: Print the shape of T
print("Shape of T:", T.shape)

# Matrix T is a transformation matrix obtained by rearranging the rows of an identity matrix. 
# It has the same shape as the original identity matrix and is used to manipulate another matrix (img_average_channel) with the same dimensions through matrix multiplication, possibly altering its values or structure.
```

## Part 6

```python
# Matrix multiplication of T and img_average_channel
transformed_image = np.dot(T, img_average_channel)

# Check the shape of the result
print("Shape ", transformed_image.shape)

# Plot transformed_image
plt.figure(figsize=(6, 6))
plt.imshow(transformed_image)  # Display in grayscale
plt.title("Transformed Image")
plt.show()

# Te shape of the result reflects the number of rows from T and the number of columns from img_average_channel, 
# indicating how the elements of img_average_channel have been transformed or linearly combined based on the rows of T.
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
import numpy as np


# Check if matrix T is invertible (has a non-zero determinant)
determinant_T = np.linalg.det(T)

if determinant_T != 0:
    print("Matrix T is invertible ")
    # T is invertible, compute its inverse
    
    Tinv = np.linalg.inv(T)

    # Perform the inverse operation to retrieve the original image
    retrieved_image = np.dot(Tinv, transformed_image)

    # Plot the retrieved original image
    plt.figure(figsize=(6, 6))
    plt.imshow(retrieved_image)  # Display in grayscale
    plt.title("Retrieved Original Image")
    plt.show()

```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image (replace 'your_image.jpg' with the actual image file path)
#image = np.array(Image.open('../../Images/lemon.jpeg'))

# Define a random permutation for rows and columns
num_rows, num_columns, _ = image.shape
row_permutation = np.random.permutation(num_rows)
column_permutation = np.random.permutation(num_columns)

# Scramble both rows and columns
scrambled_image = image[row_permutation, :, :]
scrambled_image = scrambled_image[:, column_permutation, :]

# Display the scrambled image
plt.imshow(scrambled_image)
plt.title("Scrambled Image (Both Rows and Columns)")
plt.show()

# Invert the row and column permutations
inverse_row_permutation = np.argsort(row_permutation)
inverse_column_permutation = np.argsort(column_permutation)

# Unscramble the image
recovered_image = scrambled_image[inverse_row_permutation, :, :]
recovered_image = recovered_image[:, inverse_column_permutation, :]

# Display the recovered image
plt.imshow(recovered_image)
plt.title("Recovered Image")
plt.show()

```
